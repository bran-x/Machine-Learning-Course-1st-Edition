# Balanceo de clases

```python
# Generar el dataset
from sklearn.datasets import make_classification
X1, Y1 = make_classification(n_samples = 250, n_features = 2, n_informative=2, 
                           n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=1, random_state=1)
```

```python
import matplotlib.pyplot as plt
def plot_class(X,Y, title):
  plt.title(title)
  plt.scatter(X[:,0], X[:,1], 25.0, Y)
  plt.show()
```

```python
plot_class(X1,Y1,'Clases balanceadas')
```

```python
X2, Y2 = make_classification(n_samples = 250, n_features = 2, n_informative=2, 
                           n_redundant=0, n_repeated=0, n_classes=2, 
                           n_clusters_per_class=1, random_state=1, weights=(0.8, 0.2))
```

```python
plot_class(X2,Y2,'Clases desbalanceadas')
```

```python
# Verificamos el desbalance
print("Clase 0:", X2[Y2==0].shape)
print("Clase 1:", X2[Y2==1].shape)
```

Estrategia para balancear los datos

```python
from sklearn.utils import resample
X2_res = resample(X2[Y2==1], n_samples = X2[Y2==0].shape[0], random_state = 1)
```

```python
X2_res.shape
```

```python
import numpy as np
X3 = np.r_[X2[Y2==0], X2_res]
Y3 = np.r_[Y2[Y2==0], np.ones(shape=[X2[Y2==0].shape[0]])]
```

```python
X3.shape
```

```python
# Graficamos
plot_class(X3,Y3,'Resample')
```

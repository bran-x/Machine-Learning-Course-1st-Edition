## Observamos la data

```python
import pandas as pd
train_data = pd.read_csv('train.csv')
```

```python
train_data.head()
```

```python
train_data.shape
```

```python
X_train = train_data.values[:,1:]
Y_train = train_data.values[:,0]
```

```python
# Volvemos a separar los valores
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, train_size=0.8)
```

```python
# Listas para almacenar scores y nombres
modelos = list()
scores = list()
```

# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
modelos.append('GaussianNB')
scores.append(gnb.score(x_test,y_test))
```

## Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
modelos.append('MultinomialNB')
scores.append(mnb.score(x_test,y_test))
```

## Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
modelos.append('BernoulliNB')
scores.append(bnb.score(x_test,y_test))
```

## SGDClassifer

```python
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier()
```

```python
sgd.fit(x_train,y_train)
```

```python
modelos.append('SGDClassifier')
scores.append(sgd.score(x_test,y_test))
```

## Comparamos los modelos

```python
data = {'Modelos':modelos, 'Scores':scores}
dataframe = pd.DataFrame(data=data)
```

```python
dataframe.sort_values('Scores', inplace=True, ascending=False)
```

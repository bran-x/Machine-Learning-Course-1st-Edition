# AdaBoost

## 1. Dataset

```python
from sklearn.datasets import make_classification
X,Y = make_classification(n_samples=1000, n_features=8,n_informative=8, n_redundant=0, n_repeated=0, n_classes=5, n_clusters_per_class=3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size=0.8)
```

## 2. Modelo

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier()
```

```python
abc.fit(X_train,Y_train)
```

```python
abc.score(X_train,Y_train)
```

```python
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

acc = []
for i in range(1,100):
  a = cross_val_score(AdaBoostClassifier(n_estimators=i,learning_rate=0.1),X,Y,cv=5).mean()
  acc.append(a)
plt.plot(acc)
plt.show()
```

```python
from sklearn.tree import plot_tree
plot_tree(abc.estimators_[-2], filled=True)
```

```python
from sklearn.datasets import load_iris
data = load_iris()
X = data.data
Y = data.target
```

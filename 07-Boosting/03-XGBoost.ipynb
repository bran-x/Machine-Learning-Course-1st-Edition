## 1. Dataset

```python
from sklearn.datasets import make_classification
X,Y = make_classification(n_samples=1000, n_features=8,n_informative=8, n_redundant=0, n_repeated=0, n_classes=5, n_clusters_per_class=3)
```

```python
from sklearn.model_selection import cross_val_score
```

## 2. Modelo

```python
xgbclass = xgb.XGBClassifier()
```

```python
cross_val_score(xgbclass, X,Y,cv=4)
```

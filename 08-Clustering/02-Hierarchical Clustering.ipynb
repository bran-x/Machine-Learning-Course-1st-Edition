# Clustering Jerarquico

## Dendrograma

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
X,Y = make_blobs(20, n_features=2, centers=4, random_state=1)
plt.scatter(X[:,0],X[:,1],20.0,'b')
```

Proceso para graficar Dendrograma

```python
# Distancia
from scipy.spatial.distance import pdist
Xdist = pdist(X, metric='euclidean')
```

```python
X.shape
```

```python
Xdist.shape
```

```python
# Enlace
from scipy.cluster.hierarchy import linkage
XLink = linkage(Xdist, method='ward')
```

```python
XLink
```

```python
# Dendrograma
from scipy.cluster.hierarchy import dendrogram
Xdend = dendrogram(XLink)
```

## 1. Clustering Aglomerativo

```python
# Dataset
from sklearn.datasets import make_blobs
n = 2500
X, Y = make_blobs(n_samples=n, n_features=2, centers=4, random_state=10, cluster_std=1.5)
```

```python
plt.scatter(X[:,0],X[:,1],20.0,'b')
```

Clustering

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
Yp = ac.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],20.0,Yp)
```

```python
# Evaluamos
from sklearn.metrics import silhouette_score
print('Silhouette score: {}'.format(silhouette_score(X,Yp)))
```

Clustering 2

```python
ac = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')
Yp = ac.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],20.0,Yp)
```

```python
print('Silhouette score: {}'.format(silhouette_score(X,Yp)))
```

Clustering 3

```python
ac = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')
Yp = ac.fit_predict(X)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1],20.0,Yp)
```

```python
print('Silhouette score: {}'.format(silhouette_score(X,Yp)))
```

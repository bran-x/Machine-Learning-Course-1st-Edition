# Biclustering

```python
import numpy as np
n1 = 80
n2 = 120
max_feat = 25
matrix = np.random.randint(0,max_feat, size=(n1,n2)) 
```

```python
import matplotlib.pyplot as plt
plt.imshow(matrix)
```

Biclustering Spectral

```python
from sklearn.cluster import SpectralBiclustering
sb = SpectralBiclustering(n_clusters=3, random_state=1)
```

```python
sb.fit(matrix)
```

```python
sb.row_labels_
```

```python
sb.column_labels_
```

```python
matrix_clustered = np.outer(np.sort(sb.row_labels_)+1, np.sort(sb.column_labels_)+1)
```

```python
plt.imshow(matrix_clustered)
```

```python
matrix_clustered = np.outer(sb.row_labels_+1, sb.column_labels_+1)
```

```python
plt.imshow(matrix_clustered)
```

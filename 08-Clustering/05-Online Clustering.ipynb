```python
from sklearn.datasets import make_blobs
X,Y = make_blobs(1250, n_features=2, centers= 4, random_state=1, cluster_std=1.5)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],20.0,Y)
```

## Mini-batch K-Means

```python
from sklearn.cluster import MiniBatchKMeans
batch_size=50
mbkm = MiniBatchKMeans(n_clusters=4, batch_size=batch_size, random_state=1)
```

```python
X_batch = []
Y_batch = []
for i in range(0, 1250, batch_size):
  mbkm.partial_fit(X[i:i+batch_size])
  X_batch.append(X[:i+batch_size])
  Y_batch.append(mbkm.predict(X[:i+batch_size]))
```

```python
fig, axs = plt.subplots(5,5, figsize=(12,12))
for i in range(5):
  for k in range(5):
    axs[i][k].scatter(X_batch[5*i+k][:,0],X_batch[5*i+k][:,1],1.0,Y_batch[5*i+k])
plt.show()
```

## BIRCH

```python
from sklearn.cluster import Birch
birch = Birch(n_clusters=4, threshold=0.2, branching_factor=50)
```

```python
X_batch = []
Y_batch = []

for i in range(0, 1250, batch_size):
  birch.partial_fit(X[i:i+batch_size])
  X_batch.append(X[:i+batch_size])
  Y_batch.append(birch.predict(X[:i+batch_size]))
```

```python
fig, axs = plt.subplots(5,5, figsize=(12,12))
for i in range(5):
  for k in range(5):
    axs[i][k].scatter(X_batch[5*i+k][:,0],X_batch[5*i+k][:,1],1.0,Y_batch[5*i+k])
plt.show()
```
